```python
import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Open text file 
file = open('deep_fakes_article.txt')

# Create a new text file and fill it with a list of lines from the article
list_file = open('list_file.txt', mode="a+", encoding="utf-8")
sentence_list = file.readlines()
for line in sentence_list:
    list_file.write(line)
list_file.close()
```

```python
# Open the newly created file with list of textlines
text = open('list_file.txt').read()
```

```python
# Generate the wordcloud
wordcloud = WordCloud(max_font_size=120, max_words=200, background_color="white",
                      colormap='Spectral').generate(text)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Create a mask to use as a shape for the wordcloud
mask = np.array(Image.open('oval_shape.png'))

# Update the stopwords
stopwords = set(STOPWORDS)
stopwords.update(['Hanks', 'even', 'Tom', 're', 'middle', 'Hwang', 'Tully', 'say', 's', 'lot', 'says', 
                 'Foster', 'used', 'Hankses'])
```

```python
# Final version with updated stopwords and shape
wordcloud = WordCloud(max_font_size=120, max_words=200, background_color="white", stopwords=stopwords, 
                      mask=mask, colormap='Spectral', width=mask.shape[1], height=mask.shape[0]).generate(text)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("wordcloud.png")
```
